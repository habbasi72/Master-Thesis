First step :
 Considering only Month column:
 - Creating a list of random timestamp 
 - Extracting only Month names
 - Adding some noise by cutting some characters by random in    order to simulate typos
 - Applying fuzzy string matching logic and Levenshtein algorithm to find the best match(The core is difflib library)
 

```python
import pandas as pd
import numpy as np
import random
import string
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import warnings
warnings.filterwarnings('ignore')
```

```python
#Creating a DataFrame of predefined size
size = 100
df = pd.DataFrame(index=range(size))
```

```python
#Making a column of Random Timestamps
df["Timestamp"] = [ pd.Timestamp(2017, np.random.randint(1,13), 1) for _ in range(100) ]

#Extracting just the Months from above data
df["STD_Months"] = df["Timestamp"].apply(lambda x: x.strftime('%B'))
df["Noisy_Months"] = df["Timestamp"].apply(lambda x: x.strftime('%B'))
df.head(10)
```

```python
#Adding 10% random NaNs to to the column
ix = [(row) for row in range(df["Noisy_Months"].shape[0])]
for row in random.sample(ix, int(round(.1*len(ix)))):
    df["Noisy_Months"].iat[row] = np.nan
df.head(10)
```

```python
# #Random string generator
# def id_generator(size=2, chars=string.ascii_uppercase + string.digits):
#     return ''.join(random.choice(chars) for _ in range(size))
# randomString = id_generator()
# print(randomString)
```

```python
#Replacing letters with random letter
from string import ascii_letters
for row in random.sample(ix, int(round(.4*len(ix)))):
#     s = df["Noisy_Months"].iat[row]
    if pd.notnull(df["Noisy_Months"].iat[row]):
        #Indexes
        #inds = [i for i,_ in enumerate(s) if not s.isspace()]
        indxs = [i for i , _ in enumerate(df["Noisy_Months"].iat[row])]
        print("indxs:" ,indxs)
        #Choose 3 random indexes
        sam = random.sample(indxs , 2)
        print("sam:" ,sam)
        #Choose 3 random character
        letts =  iter(random.sample(ascii_letters, 2))
        #make the string a list
        lst = list(df["Noisy_Months"].iat[row])
        for ind in sam:
            lst[ind] = next(letts)
    df["Noisy_Months"].iat[row]=("".join(lst))
    print("row:" ,row)    
    print("s:" ,df["Noisy_Months"].iat[row])
   
```

```python
#Drob fields containing missing values using Pandas's DropNA method
df.dropna(inplace=True)
df.shape
df.head(10)
```

```python
#Applying Fuzzy Levenshtein string matching 
# months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November" ,"December"   ] 
choices_months = df["STD_Months"].unique()
%timeit df["Fuzzy_Info"] = df["Noisy_Months"].apply(lambda x: process.extract(x, choices_months, limit=1)[0])
df["Best_Match"] = df["Noisy_Months"].apply(lambda x: process.extractOne(x, choices_months)[0])
df.head(10)
```

```python
#Calculating the Accuracy
from sklearn.metrics import accuracy_score
print("Accuracy is:",accuracy_score(df["STD_Months"] , df["Best_Match"]))
```
