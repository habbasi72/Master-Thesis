```python
import pandas as pd
m1 = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November" ,"December"] 
m2 = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November" ,"December"] 


def CommonLetters(s1, s2):
    l1=list(''.join(s1.split()))
    l2=list(''.join(s2.split()))
    return [x for x in l1 if x in l2]
cols=('Month', 'Total')

df = pd.DataFrame(index=range(12),columns=cols)
total=0

for i,x in enumerate(m1):
    for j,y in enumerate(m2):
        month1 = x
        month2 = y
#         print(month1 ,'-', month2,'->' , CommonLetters(month1,month2))        
#         print('#of common letters: ', len(CommonLetters(month1,month2)) , '\n')
        total += len(CommonLetters(month1,month2))
#     print('Total: ' , total)
#     print('//////////')
    df['Month'].iat[i] = month1
    df['Total'].iat[i] = total
```
