Q: How to deal with the cases of having equal edit distances?
    -Choosing 1 at random

```python
import jellyfish
#Levenshtein Distance

months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November" ,"December"   ] 
ix = [(row) for row in range(df['10_Percent'].shape[0])]
for row in ix:
    print("row: " , row)
    ld_best = 20
    best_match = ''
    for month in months:
        s1 = df['10_Percent'].iat[row]
        print("s1: " , s1 )
        s2 = month        
        print("s2: " , s2)
        ld = jellyfish.levenshtein_distance(s1 , s2)
        print("LD :" , ld)
        if ld < ld_best:
            ld_best = ld
            best_match = s2
        df['10P_LD'].iat[row] = best_match
        print('\n')
    print("Best Match: " , best_match)
    print("ld_best: " , ld_best)        
    print("*******")
#     df['10P_LD'].append(s2)
```
