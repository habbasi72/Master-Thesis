```python
import pandas as pd
df1=pd.read_csv('Noisy_Data.csv')
df=df1.drop('Unnamed: 0' , 1)
df
```

Damerau-Levenshtein Distance
A modification of Levenshtein distance, Damerau-Levenshtein distance counts transpositions (such as ifhs for fish) as a single edit.

```python
import jellyfish
#Damerau-Levenshtein Distance    
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November" ,"December"   ] 
colnames=list(df.columns.values)
for x in colnames:
    if x!='STD_Months' and x!='Date':
        col=x
        print(col)
        df[col+'_Damerau'] = ''
        ix = [(row) for row in range(df[col].shape[0])]
        for row in ix:
            print("row: " , row)
            Dld_best = 20
            best_match = ''
            for month in months:
                s1 = df[col].iat[row]
                print("s1: " , s1 )
                s2 = month        
                print("s2: " , s2)
                Dld = jellyfish.damerau_levenshtein_distance(s1 , s2)
                print("Damerau LD :" , Dld)
                if Dld < Dld_best:
                    Dld_best = Dld
                    best_match = s2
                df[col+'_Damerau'].iat[row] = best_match
                print('\n')
            print("Best Match: " , best_match)
            print("Dld_best: " , Dld_best)        
            print("*******")       
```

```python
#Damerau-Levenshtein Accuracy
from sklearn.metrics import accuracy_score

NameOfColumns = list(df)
DLdAccuracy = []
for x in NameOfColumns:
    if "Damerau" in x:
        print("LDD Accuracy : " , accuracy_score(df["STD_Months"] , df[x])) 
        DLdAccuracy.append(accuracy_score(df["STD_Months"] , df[x]))
print(DLdAccuracy)
```
