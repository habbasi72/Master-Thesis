```python
import pandas as pd
df1=pd.read_csv('Noisy_Data.csv')
df=df1.drop('Unnamed: 0' , 1)
```

```python
# Creating an empty column
# df['100P_LD'] = ''
```

```python
import jellyfish
#Levenshtein Distance for EQUAL Lenghts    
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November" ,"December"   ] 
colnames=list(df.columns.values)
for x in colnames:
    if x!='STD_Months' and x!='Date':
        col=x
        print(col)
        df[col+'_LD'] = ''
        ix = [(row) for row in range(df[col].shape[0])]
        for row in ix:
            print("row: " , row)
            ld_best = 20
            best_match = ''
            for month in months:
                s1 = df[col].iat[row]
                print("s1: " , s1 )
                s2 = month        
                print("s2: " , s2)
                if len(s1)==len(s2):
                    ld = jellyfish.levenshtein_distance(s1 , s2)
                    print("LD :" , ld)
                    if ld < ld_best:
                        ld_best = ld
                        best_match = s2
                df[col+'_LD'].iat[row] = best_match
                print('\n')
            print("Best Match: " , best_match)
            print("ld_best: " , ld_best)        
            print("*******")       
```

```python
#Levenshtein Accuracy
from sklearn.metrics import accuracy_score

NameOfColumns = list(df)
ELLdAccuracy = []
for x in NameOfColumns:
    if "LD" in x:
        print("LD Accuracy : " , accuracy_score(df["STD_Months"] , df[x])) 
        ELLdAccuracy.append(accuracy_score(df["STD_Months"] , df[x]))
print(ELLdAccuracy)
```
