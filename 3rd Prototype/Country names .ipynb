```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

#Import and read the csv file
df = pd.read_csv('countries.csv')
df
```

```python
#Cleaning the country names(Removing additional info and Stop words) 
df['cleanedNames'] = ''
df['length'] = ''
ix = [(row) for row in range(df['Names'].shape[0])]
sep1 = ','
sep2 = '('
stopwords = ['a' , 'the' , 'and' , 'of']
for row in ix:
    ctry = df['Names'].iat[row]
    ctry = ctry.split(sep1 , 1)[0]
    ctry = ctry.split(sep2 , 1)[0]
    querywords = ctry.split()
    resultwords  = [word for word in querywords if word.lower() not in stopwords]
    df['cleanedNames'].iat[row] = ' '.join(resultwords)
    df['length'].iat[row] = len(df['cleanedNames'].iat[row])
df 
```

```python
Values = df['length'].value_counts()
print(Values)
Values.plot.bar()
plt.xlabel('Length')
plt.ylabel('Number')
plt.title('(Length Distribution)' , fontsize=14)
plt.yticks(np.arange(0 , 50, 2))
plt.grid(color='b', linewidth=0.3)
plt.show()
```

```python
# Distribution plot
sns.distplot(Values)
plt.show()
```
