```python
import pandas as pd
import numpy as np
import random
import string
from decimal import Decimal , ROUND_HALF_UP
import matplotlib.pyplot as plt

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#Import dataset
df=pd.read_csv('Initial.csv')
df=df.drop('Unnamed: 0' , 1)
df
```

```python
Values = df['STD_Months'].value_counts()
print(Values)
Values.plot.barh()
plt.show()
```

```python
#Make a copy of initial column
df['10_Percent'] = df['STD_Months']
#Creating 10% noise
p1 = 0.1
#Extract the column's indeces [0,1,...,99]
ix1 = [(row) for row in range(df['10_Percent'].shape[0])]
for row in ix1:
    print("row:" ,row)
    indxs1 = [i for i , _ in enumerate(df['10_Percent'].iat[row])]
    print(df['10_Percent'].iat[row])
    print ('Index:',indxs1)
    length1 = len(indxs1)
    print('length:',length1)
    letters = Decimal(str(length1*p1)).quantize(Decimal("1"), rounding=ROUND_HALF_UP)
    positions1 = random.sample(range(length1) , int(letters))
    print('positions:',positions1)
    lst1 = list(df['10_Percent'].iat[row])
    for ind1 in positions1:
        lst1[ind1]=random.choice(string.ascii_lowercase)
        df['10_Percent'].iat[row]=("".join(lst1))
    print("s:" ,df['10_Percent'].iat[row])
    print('***')
```
